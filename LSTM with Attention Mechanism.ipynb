```python
# Problem with Long Sequences 
# Encoder-Decoder RNN: one set of LSTMs learn to encode input sequences into
#    a fixed length internal representation, and second set of LSTMs read the 
#    internal representation and decode it into an output sequence

# It achieves excellent results on a wide range of problems. But it suffers from 
#    the constraint that all input sequences are forced to be encoded to a fixed
#    length internal vector
# This cause limit of performance of these networks, especially when having 
#    long input sequences, such as very long sentences in text translation problems

# Potential issues with encoder-decoder approach is that a neural network needs
# to be able compress all the necessary information of a source sentence into a 
# fixed length vector. 
# This may take it difficult for the neural network to cope with long sentences, 
# especially those that are longer than the sentences in the training corpus

Image("example_encoder_decoder.png", width=500)
```

```python
# Attention within Sequences

# Attention is the idea of freeing the encoder-decoder arch. from the fixed-length internal
# representation

# This is done by keeping the intermediate outputs from the encoder LSTM from 
# each step of the input sequence and training the model to learn to pay 
# selective attention to these inputs and relate them to items in the output sequence.

# Each item in the output sequence is conditional on selective items in the 
# input sequence

# Everytime the proposed model generates a word in a translation, it searches for 
# a set of positions in a source sentence, where the most relevant information is concentrated
# Then it will predict a word based on the context vectors associated with these source positions 
# and all the previous generated words

# Instead of squashing all the informations of source sentence into a fixed-length vector, 
# it encodes the input sentence into a sequence of vectors and choose a subset of these vector 
# adaptively while decoding the translation

# Its computational expensive model, but the results is better 

# Proposed way is by providing an intuitive way to inspect the alignment between the words
# in a generated translations and those in a source sentence. This is done by 
# visualizing the annotation weights

# Each row of a matrix in each plot indicates the weights associated with the annotations. 
# From this we see which positions in the source sentence were considered more important 
# when generating the target word.

# Basically, having attention mechanism in RNN helps the algorithm to decide which 
# part of the sentence to be focused on instead of squashing the whole text to fit 
# the fixed-length vector



image1 = Image("attention-1.png", width=800)
image2 = Image("attention-2.png", width=800)
display(image1, image2)
```
